```python
#4
def shot_accuracy(shots_on_target, total_shots):
  return shots_on_target / total_shots

shot_accuracy(3,7)
```

```python
#5
def is_high_scorer(goals):
  if goals >= 15:
    return True
  else:
    return False

is_high_scorer(17)
```

```python
#6
def goalc_per_90(goals, assists, minutes):
  return (goals + assists) / (minutes / 90)

goalc_per_90(1, 2, 88)
```
