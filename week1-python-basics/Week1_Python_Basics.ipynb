```python
# Multiple values in dictionary
player_stats = [
    {"name": "Palmer", "goals": 15, "assists": 7, "minutes": 2800},
    {"name": "Neto", "goals": 6, "assists": 12, "minutes": 2500},
    {"name": "Delap", "goals": 9, "assists": 2, "minutes": 1900},
    {"name": "Joao Pedro", "goals": 12, "assists": 6, "minutes": 2700}
]

def contributions_per_90(name, goals, assists, minutes):
  total = goals + assists
  return f"{name} contributions per 90 = {round((total / (minutes / 90)), 2)}"

for p in player_stats:
  print(contributions_per_90(
      p["name"],
      p["goals"],
      p["assists"],
      p["minutes"]
  ))
```

```python
# Rewriting function so it accepts entire dictionary at once
def contributions_per_90(player):
  total = player["goals"] + player["assists"]
  minutes_per_90 = player["minutes"] / 90
  return f"{player['name']} contributions per 90 = {round((total / minutes_per_90), 2)}"

# print(contributions_per_90(player_stats[0]))

for p in player_stats:
  print(contributions_per_90(p))

```

```python
# Storing results
from pprint import pprint

results = []

for p in player_stats:
  gc90 = (p["goals"] + p["assists"]) / (p["minutes"] / 90)
  results.append({"name": p["name"], "gc90": round(gc90, 2)})

# pprint(results)
# for r in results:
#   print(r)

sorted_results = sorted(results, key=lambda x: x["gc90"], reverse=True)
for r in sorted_results:
  print(f"{r['name']} - {r['gc90']} contributions per 90")
```
