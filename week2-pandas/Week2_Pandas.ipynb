```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/dmnarula/football-ds-learning/refs/heads/main/sample_data/pl_24_25_small.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.head()
df.info()
df.describe().round(2)
```

```python
df[["name", "club", "goals", "assists"]].head()
```

```python
df[df["goals"] >= 15]
```

```python
df["goal_contributions"] = df["goals"] + df["assists"]
df.head()
```

```python
df["gc_per_90"] = (df["goal_contributions"] / (df["minutes"] / 90)).round(2)
df.head()
```

```python
df.sort_values(by="gc_per_90", ascending=False).head(10).round(2)
```

```python
club_stats = df.groupby("club")[["goals", "assists", "goal_contributions"]].sum()
club_stats.sort_values(by="goal_contributions", ascending=False)
```

```python
club_avg = df.groupby("club")[["gc_per_90", "xG", "xA"]].mean().round(2)
club_avg.sort_values(by="xG", ascending=False)
```

```python
# Practice
club_avg_minutes = df.groupby("club")[["minutes"]].mean().round(2)
club_avg_minutes.sort_values(by="minutes", ascending=False)
```

```python
df["xGI"] = df["xG"] + df["xA"]
df["xGI_per_90"] = (df["xGI"] / (df["minutes"] / 90)).round(3)
df.head()
```

```python
# Practice
df.sort_values(by="xGI_per_90", ascending=False).head(10)
```

```python
df[(df["goals"] >= 10) & (df["xGI_per_90"] >= 0.6)]
```

```python
df[(df["club"] == "Arsenal") | (df["gc_per_90"] >= 0.7)]
```

```python
# Practice
df[(df["club"] == "Liverpool") | (df["xGI_per_90"] >= 0.7)]
```

```python
clubs = ["Arsenal", "Chelsea", "Liverpool", "Tottenham"]

filtered = df[
    (df["club"].isin(clubs)) &
    (df["gc_per_90"] >= 0.6)
]

filtered[["name", "club", "gc_per_90", "xGI_per_90"]].sort_values(
    by="gc_per_90", ascending=False
)
```

```python
# Practice
clubs_ex = ["Arsenal", "Man City", "Liverpool"]

filtered_ex = df[
    (df["club"].isin(clubs_ex)) &
    (df["xG"] >= 10)
]

filtered_ex[["name", "club", "xG"]].sort_values(
    by="xG", ascending=False
)
```

```python
recruitment_table = df[
    (df["minutes"] >= 900) & # 10 games min
    (df["xGI_per_90"] >= 0.4) # decent attacking
][["name", "club", "minutes", "gc_per_90", "xGI_per_90"]]

recruitment_table.sort_values(
    by="xGI_per_90", ascending=False
).head(15)
```

```python
recruitment_table_clean = recruitment_table.sort_values(
    by="xGI_per_90", ascending=False
).reset_index(drop=True)

recruitment_table_clean.head(15)
```

```python
# Practice
def_table = df[
    (df["minutes"] >= 2000)
][["name", "club", "minutes", "xG", "xA"]]

def_table_clean = def_table.sort_values(
    by="minutes", ascending=False
).reset_index(drop=True)

def_table_clean.head(15)
```

```python
# Saving recruitment table as CSV
recruitment_table_clean.to_csv("recruitment_table.csv", index=False)
```

```python
# Saving def table as CSV
def_table_clean.to_csv("def_table.csv", index=False)
```
